```python
#Queestion 5 using regression models - statsmodel OLS
X=df[['Dist']]
y=df[['CPL']]
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
print(est.fit().f_pvalue)
est2=est.fit()
print(est2.summary())
```

```python
#Question 6 comparing models
X=df[['Dist','Wgt']]
y=df[['CPL']]
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
print(est.fit().f_pvalue)
est2=est.fit()
print(est2.summary())
```

```python
#Question 6 - First adding column for Distance ^2 and checking
df['DistSqr']=df['Dist']**2
df.head()
```

```python
#Question 6 - running OLS regression with the new independent variable
X=df[['Dist','DistSqr']]
y=df[['CPL']]
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
print(est.fit().f_pvalue)
est2=est.fit()
print(est2.summary())
```

```python
#Question 7 distance and weight product
df['DistWgt']=df['Dist']*df['Wgt']
df.head()
```
